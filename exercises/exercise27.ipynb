*A dog swims from one bank of a river to his master on the other bank. It always swims in the direction its master. Suppose the speed of the river is 1, and the dog’s speed is 2. The dog starts at point (1, 0) and its master waits at point (0, 0).*

$\textrm{(i)} \quad$ *Draw the dog’s approximate trajectory by constructing the piecewise linear path corresponding to* $x^i = 1−0.1i, i = 0, \dots, 10.$ *The dog’s direction of motion at the starting point* $(x^0, y^0)$ *is given by addition of the velocity vectors. One obtains an approximate value* $y^1$ *at the point* $x^1$*. One analogously constructs a point* $(x^2, y^2)$ *from* $(x^1, y^1),$ *etc.*

```octave
%plot -s 1280,720

pos_master = [0; 0];
pos_dog = [1;0];
spd_dog = 2;
vel_river = [0; 1];

N = 10;
for i = 1 : N
    line = pos_master - pos_dog(:,i);
    vel_dog = vel_river + spd_dog * line / norm(line);
    pos_dog(:,i+1) = pos_dog(:,i) + 1/N * vel_dog / abs(vel_dog(1));
endfor

plot(pos_dog(1,:), pos_dog(2,:))
title("approximate trajectory");
xlabel("x");
ylabel("y");
```

$$ $$
$\textrm{(ii)} \quad$ *The dog’s path is described by the equations*
$$\begin{cases}
x'(t) = - \frac{2x}{\sqrt{x^2+y^2}} \\
y'(t) = 1 - \frac{2y}{\sqrt{x^2+y^2}}
\end{cases}$$
*Compute the trajectory.*

```octave
function vel = F(t, pos)
    vel = [0; 1] - 2 * pos / norm(pos);
endfunction

pos = [1; 0];
N = 1000;
t = linspace(0, 1, N + 1);

for i = 1 : N
    h = t(i+1) - t(i);
    pos(:,i+1) = pos(:,i) + h * F(t(i), pos(:,i));
endfor

plot(pos(1,:), pos(2,:))
title("trajectory");
xlabel("x");
ylabel("y");
```

$$ $$ $\textrm{(iii)} \quad$ *How long does it take for the dog to reach its master?*

```octave
hold on;
plot(t, pos(1,:))
plot(t, pos(2,:))
legend("x coordinate", "y coordinate");
xlabel("time");
title("displacement");
```

```octave
[val, indx] = max(pos(1,:) < 0);
[val, indy] = max(pos(2,:) < 0);
t(max(indx, indy))
```

At approximately time 0.668 the dog reaches his master.
